```python
import pandas as pd
import numpy as np
import seaborn as sns



# Read the dataset
df = pd.read_csv('city_day.csv')
#Task a: Retrieve data from ‘AQI_IND.csv’ in the form of data frame and check the number of rows and columns in the dataset.

# Check the number of rows and columns
print("Number of Rows: ", df.shape[0])
print("Number of Columns: ", df.shape[1])


#Task b: Find the columns and the associated data types, as well as whether or not they include null values.

# Find the columns and their data types
print(df.info())

# Check if there are any null values in the dataset
print("\nNull Values in the dataset: ", df.isnull().sum().sum())


#Task c: Obtain statistical summary for all the columns in the dataset.
# Get the statistical summary for all columns
print(df.describe())


#Task d: Explore the possibility of correlation between the variables. Visualize using heatmap.

# Compute the correlation matrix
corr = df.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

#Task e



import matplotlib.pyplot as plt



# Selecting only the required columns
cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']

# Describing the dataset
desc_df = df[cols].describe()

# Printing the description
print(desc_df)

# Visualizing the data distribution
fig, ax = plt.subplots(nrows=4, ncols=3, figsize=(20,20))
for i, col in enumerate(cols):
    sns.histplot(df[col], kde=True, ax=ax[i//3, i%3])
    ax[i//3, i%3].set_title(col)

# Checking the skewness of the features
skewness_df = df[cols].skew(axis=0, skipna=True)

# Printing the skewness values
print("\nSkewness of the features:\n")
print(skewness_df)



```
